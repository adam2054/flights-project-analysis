1. Distribution of passenger numbers over the months:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='month', y='passengers', data=df_flights)
plt.title('Verteilung der Passagierzahlen √ºber die Monate')
plt.xlabel('Monat')
plt.ylabel('Passagiere')
plt.show()
```

2. Development of passenger numbers over the years:

```python
plt.figure(figsize=(10, 7))
sns.lineplot(x='year', y='passengers', data=df_flights)
plt.title('2. Development of passenger numbers over the years:')
plt.xlabel('year')
plt.ylabel('passengers')
plt.show()
```

```python
# Create pivot table to put the data into the format required for the heatmap

pivot_passengers = df_flights.pivot_table(values='passengers', index='month', columns='year')
pivot_passengers
```

3. Heatmap of passenger numbers per month and year:

```python

#  Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_passengers, cmap='YlGnBu', annot=True, fmt='.1f')
plt.title('Heatmap of the total passenger numbers per month per year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

Passenger counts for the years 1949 and 1960

```python

# Passenger counts for the years 1949 and 1960
passengers_1949 = 148  # Assume 148 passengers forjan in the year 1949 (as mentioned)
passengers_1960 = 622  # Passengers for the peak year 1960 (as mentioned)

# Calculate percentage increase
percentage_increase = ((passengers_1960 - passengers_1949) / passengers_1949) * 100

# Data for the bar plot
years = ['1949', '1960']
passenger_counts = [passengers_1949, passengers_1960]

# Create a bar plot to visualize the growth
plt.figure(figsize=(8, 6))
plt.bar(years, passenger_counts, color=['blue', 'orange'])
plt.ylabel('Passenger Counts')
plt.title('320% Growth in Passenger Numbers (Jan-1949 to Dec-1960)')
plt.text(0, passengers_1949 + 10, f'{passengers_1949}', ha='center')
plt.text(1, passengers_1960 + 10, f'{passengers_1960}\n+{percentage_increase:.0f}%', ha='center')
plt.ylim(0, passengers_1960 + 50)  # Adjust y-axis limit for better visualization
plt.show()
```
